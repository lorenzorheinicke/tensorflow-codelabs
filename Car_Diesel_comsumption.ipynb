```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

## Define and Compile the Neural Network

```
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

## Train the Network

```
model.fit(xs, ys, epochs=500)
```

## Predict

```
print(model.predict([10.0]))
```

## Provide Fuel consumption data to model

```
#x_liters = np.array([49.91, 50.06,49.57, 50.51, 50.51, 49.11], dtype=float)
#y_kms = np.array([246.7, 370.4, 349.5, 331.4, 377.6, 335.8], dtype=float)

x_liters = np.array([37.55, 49.11, 49.57, 50.06, 50.51], dtype=float)
y_kms = np.array([246.7, 335.8,349.5, 370.4, 377.6], dtype=float)
```

## Make copy of model

```
#from keras import optimizers

# All parameter gradients will be clipped to
# a maximum norm of 1.
sgd = keras.optimizers.SGD(lr=0.05, clipnorm=1.)
```

```


fuel_consumption_model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
fuel_consumption_model.compile(optimizer=sgd, loss='mean_squared_error')
```

## Train the Fuel Consumption Model

```
fuel_consumption_model.fit(x_liters, y_kms, epochs=500)
```

```
print(fuel_consumption_model.predict([51.0]))
```
